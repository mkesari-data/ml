```python
# prompt: build sample model to demonstrate the k-model clustering algorithm

import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create a sample dataset
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Create a KMeans model with 4 clusters
model = KMeans(n_clusters=4)

# Fit the model to the data
model.fit(X)

# Predict the cluster assignments for the data
y_pred = model.predict(X)

# Plot the data and the cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python

# Create a KMeans model with 4 clusters
model = KMeans(n_clusters=6)

# Fit the model to the data
model.fit(X)

# Predict the cluster assignments for the data
y_pred = model.predict(X)

# Plot the data and the cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
# prompt: generate elbowl to identify correct cluster

inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
